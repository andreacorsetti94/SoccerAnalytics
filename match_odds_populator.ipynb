# Match Odds Populator

This module will iterate over 'Match' table and will extract every odd for each match.
For any given match, after retrieving every odd, this module calculates the max and average odds for each outcome:
home win, draw, away win.

Starting from the average odds of the bookmakers, the module will also compute the 'true' odds, removing the books margin
and from that, it will store the respective average chances for the events: home win, draw and away win.

All these informations will be stored in the 'MatchOdds' table.

```python
import nbimporter
import db_utils
```

```python
def true_odds(home_odds, draw_odds, away_odds):
    margin = (1/home_odds + 1/draw_odds + 1/away_odds) - 1
    return round(home_odds*(1+margin),2), round(draw_odds*(1+margin),2), round(away_odds*(1+margin),2)
```

```python
def odds_to_chance(*odds):
    chances = []
    for odd in odds:
        chances.append(1/odd)
    return chances
```

```python
def max_avg_odds(*betting_odds):
    valid_home_odds = max_home_odd = sum_home_odd = 0
    valid_draw_odds = max_draw_odd = sum_draw_odd = 0
    valid_away_odds = max_away_odd = sum_away_odd = 0
    
    for betting_odd in betting_odds:
        home = betting_odd[0]
        draw = betting_odd[1]
        away = betting_odd[2]
        
        if(home != None):
            max_home_odd = max(home, max_home_odd)
            valid_home_odds += 1
            sum_home_odd += home
        if(draw != None):
            max_draw_odd = max(draw, max_draw_odd)
            valid_draw_odds += 1
            sum_draw_odd += draw
        if(away != None):
            max_away_odd = max(away, max_away_odd)
            valid_away_odds += 1
            sum_away_odd += away
    
    if ( valid_home_odds == 0 or valid_draw_odds == 0 or valid_away_odds == 0): return None
    
    max_odds = max_home_odd, max_draw_odd, max_away_odd
    avg_odds = round(sum_home_odd/valid_home_odds, 2), round(sum_draw_odd/valid_draw_odds,2), round(sum_away_odd/valid_away_odds,2)
    return max_odds, avg_odds
```

```python
select_sql = 'select id, B365H,B365D,B365A,BWH,BWD,BWA,IWH,IWD,IWA,LBH,LBD,LBA,PSH,PSD,PSA,WHH,WHD,WHA,SJH,SJD,SJA,VCH,VCD,VCA,GBH,GBD,GBA,BSH,BSD,BSA from Match'
insert_sql = 'INSERT INTO MATCHODDS VALUES (?,?,?,?,?,?,?,?,?,?)'
db_file = 'database.sqlite'

matches = db_utils.query(db_file,select_sql,())
for match in matches:
    match_id = match[0]
    
    B365 = match[1], match[2], match[3]
    BW = match[4], match[5], match[6]
    IWH = match[7], match[8], match[9]
    LBH = match[10], match[11], match[12]
    PSH = match[13], match[14], match[15]
    WHH = match[16], match[17], match[18]
    SJH = match[19], match[20], match[21]
    VCH = match[22], match[23], match[24]
    GBH = match[25], match[26], match[27]
    BSH = match[28], match[29], match[30]
    
    max_avg_betting_odds = max_avg_odds(B365,BW,IWH,LBH,PSH,WHH,SJH,VCH,GBH,BSH)
    if( max_avg_betting_odds == None ):
        print('WARN: Match with id',match_id,' doesnt have valid odds')
        continue
    
    max_odds = max_avg_betting_odds[0]
    avg_odds = max_avg_betting_odds[1]
    
    avg_home_odds = avg_odds[0]
    avg_draw_odds = avg_odds[1]
    avg_away_odds = avg_odds[2]
    
    max_home_odds = max_odds[0]
    max_draw_odds = max_odds[1]
    max_away_odds = max_odds[2]
    
    true_avg_odds = true_odds(avg_home_odds, avg_draw_odds, avg_away_odds)
    avg_chances_from_odds = odds_to_chance(true_avg_odds[0],true_avg_odds[1],true_avg_odds[2])
    
    avg_home_chance = round(avg_chances_from_odds[0],2)
    avg_draw_chance = round(avg_chances_from_odds[1],2)
    avg_away_chance = round(avg_chances_from_odds[2],2)
    
    db_utils.write(db_file, insert_sql, (match_id,avg_home_odds,avg_draw_odds,avg_away_odds,max_home_odds,max_draw_odds,
                               max_away_odds,avg_home_chance,avg_draw_chance,avg_away_chance))
```
